# 1. How many distinct (different) actors' last names are there?

```python
ln_query= 'select count(distinct last_name) as "nº of last names" from actor;'
ln_data = pd.read_sql_query(ln_query, engine)
display(ln_data)
```

# 2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.

```python
week_query= '''SELECT *,
CASE
WHEN date_format(convert(rental_date,date), "%a") in ('Sat', 'Sun') then 'weekend'
ELSE 'weekday'
END AS 'day_type'
FROM rental;
'''
week_data=pd.read_sql_query(week_query, engine)
week_data.head(25)
```

# 3. Get all films with ARMAGEDDON in the title.

```python
arm_query= '''SELECT title FROM film
WHERE title LIKE "%%ARMAGEDDON%%";
'''
arm_data = pd.read_sql_query(arm_query, engine)
display(arm_data)
```

# 4. Get 10 the longest films.

```python
long_query= '''SELECT title, length FROM film
ORDER BY length DESC
LIMIT 10;'''
long_data = pd.read_sql_query(long_query, engine)
display(long_data)
```

# 5. How many films include Behind the Scenes content?

```python
bts_query= '''SELECT COUNT(distinct film_id) AS num_films FROM film
WHERE special_features IN ('Behind the Scenes');'''
bts_data = pd.read_sql_query(bts_query, engine)
display(bts_data)
```

# 6. Which kind of movies (rating) have a mean duration of more than two hours?

```python
rating_query= '''SELECT rating, round(AVG(length),2) AS 'mean length' FROM film
GROUP BY rating
HAVING round(AVG(length),2)>120;'''
rating_data = pd.read_sql_query(rating_query, engine)
display(rating_data)
```

# 7. Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.

```python
len_query= '''SELECT title, length, DENSE_RANK() OVER (ORDER BY length) AS 'rank' FROM film
WHERE (length IS NOT NULL) AND NOT (length = '  ') AND NOT (length='');'''
len_data = pd.read_sql_query(len_query, engine)
display(len_data)
```
